```python
import random

target_sentence = "Hello, how are you?"
gene_pool = ",!?abcdefhigklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ "

population_size = 10

def generate_chromosome(length):
    genes = []
    while len(genes) < length:
        genes.append(gene_pool[random.randrange(0, len(gene_pool))])
    return ''.join(genes)

def calculate_fitness(chromosome):
    fitness = 0
    ctr = 0
    for i in chromosome:
        if i == target_sentence[ctr]:
            fitness += 1
        ctr += 1
    return fitness

def mutate(chromosome):
    index_to_mutate = random.randrange(0, len(chromosome))
    gene = list(chromosome)
    mutated_gene = gene_pool[random.randrange(0, len(gene_pool))]
    gene[index_to_mutate] = mutated_gene
    return ''.join(gene)

population = []
for i in range(population_size):
    population.append(generate_chromosome(len(target_sentence)))

population_fitness = []
for chromosome in population:
    population_fitness.append(calculate_fitness(chromosome))

for generation in range(10000):
    parent_index = population_fitness.index(max(population_fitness))
    parent = population[parent_index]
    child = mutate(parent)

    print("parent: ", parent)
    print("child: ", child)

    child_fitness = calculate_fitness(child)
    print("Child Fitness: ", child_fitness)

    index_to_replace = population_fitness.index(min(population_fitness))
    del population[index_to_replace]
    del population_fitness[index_to_replace]

    population.append(child)
    population_fitness.append(child_fitness)

    print("current population: ", population)
    print("current population fitness: ", population_fitness)

    if child == target_sentence:
        print("Target sentence found at Generation: ", generation)
        break

```

```python
import random

target_sentence = "Hello, how are you? I love this world!"
gene_pool = ",!?abcdefhigklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ "

population_size = 100

def generate_chromosome(length):
    genes = []
    while len(genes) < length:
        genes.append(gene_pool[random.randrange(0, len(gene_pool))])
    return ''.join(genes)

def calculate_fitness(chromosome):
    fitness = 0
    for i in range(len(chromosome)):
        if i < len(target_sentence) and chromosome[i] == target_sentence[i]:
            fitness += 1
    return fitness

def mutate(chromosome, mutation_rate=0.02):
    gene = list(chromosome)
    for i in range(len(gene)):
        if random.random() < mutation_rate:
            gene[i] = random.choice(gene_pool)
    return ''.join(gene)

def crossover(parent1, parent2):
    split = random.randint(1, len(parent1) - 2)
    return parent1[:split] + parent2[split:]

population = [generate_chromosome(len(target_sentence)) for _ in range(population_size)]
population_fitness = [calculate_fitness(chromo) for chromo in population]

for generation in range(10000):
    parent1 = random.choice(population)
    parent2 = random.choice(population)

    child = crossover(parent1, parent2)
    child = mutate(child, mutation_rate=0.02)
    child_fitness = calculate_fitness(child)

    print("Generation:", generation)
    print("Parent1: ", parent1)
    print("Parent2: ", parent2)
    print("Child:   ", child)
    print("Child Fitness: ", child_fitness)

    worst_index = population_fitness.index(min(population_fitness))
    population[worst_index] = child
    population_fitness[worst_index] = child_fitness


    if child == target_sentence:
        print("Target sentence found at Generation:", generation)
        break
```
